```python
import pandas as pd
import matplotlib.pyplot as plt

home_away = pd.read_csv(
    r"../data/overwiev__results2024-202591_home_away.csv"
)

home_away.head()
```

```python
home_away = home_away.rename(columns={
    "Squad": "team"
})

home_away["team"] = home_away["team"].str.strip()

home_away.head()
```

```python
home_away["home_advantage"] = (
    home_away["Home_Pts"] -
    home_away["Away_Pts"]
)
```

```python
home_rank = home_away.sort_values(
    by="home_advantage",
    ascending=False
)

home_rank[[
    "team",
    "Home_Pts",
    "Away_Pts",
    "home_advantage"
]]
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    home_rank["team"],
    home_rank["home_advantage"]
)

plt.xticks(rotation=45)
plt.ylabel("Home Advantage (Points)")
plt.title("Home Advantage Across Teams")

plt.show()
```

```python
home_away["goal_diff_home_away"] = (
    home_away["Home_GF"] -
    home_away["Away_GF"]
)
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    home_away["team"],
    home_away["goal_diff_home_away"]
)

plt.xticks(rotation=45)
plt.ylabel("Goal Difference (Home vs Away)")
plt.title("Scoring Difference: Home vs Away")

plt.show()
```
