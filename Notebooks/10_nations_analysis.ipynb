```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

nations = pd.read_csv(
    r"../data/overwiev__nations.csv"
)

nations.head()
```

```python
nations.columns
```

```python
nations = nations.rename(columns={
    "Nation": "nation",
    "# Players": "player_count",
    "Min": "minutes_played",
    "List": "players"
})

nations.head()
```

```python
top_nations = nations.sort_values(
    by="player_count",
    ascending=False
)

top_nations[[
    "nation",
    "player_count"
]].head(10)
```

```python
top_nations["nation"] = top_nations["nation"].astype(str)
top_nations = top_nations.dropna(subset=["nation", "player_count"])
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    top_nations["nation"].head(10),
    top_nations["player_count"].head(10)
)

plt.xticks(rotation=45)
plt.ylabel("Number of Players")
plt.title("Top Nationalities in the Premier League")

plt.show()
```
