```python
players = players.rename(columns={
    "Player": "player",
    "Squad": "team",
    "Performance_Gls": "goals",
    "Performance_Ast": "assists",
    "Performance_G+A": "goal_contributions",
    "Per 90 Minutes_Gls": "goals_per_90",
    "Per 90 Minutes_Ast": "assists_per_90",
    "Per 90 Minutes_G+A": "g+a_per_90",
    "Playing Time_90s": "matches_equivalent"
})
```

```python
players = players[players["matches_equivalent"] >= 5]
```

```python
players = players[players["matches_equivalent"] >= 5]
```

```python
top_efficiency = players.sort_values(
    by="goals_per_90",
    ascending=False
)[["player", "team", "goals_per_90"]]

top_efficiency.head(10)
```

```python
top_contributors = players.sort_values(
    by="g+a_per_90",
    ascending=False
)[["player", "team", "g+a_per_90"]]

top_contributors.head(10)
```

```python
top_contributors = players.sort_values(
    by="g+a_per_90",
    ascending=False
)[["player", "team", "g+a_per_90"]]

top_contributors.head(10)
```

```python
team_goal_totals = players.groupby("team")["goals"].sum().reset_index()

team_goal_totals.head()
```

```python
players = players.merge(
    team_goal_totals,
    on="team",
    suffixes=("", "_team_total")
)

players.head()
```

```python
players["team_dependency_ratio"] = (
    players["goals"] / players["goals_team_total"]
)

players.head()
```

```python
players_filtered = players[players["goals"] >= 5]

dependency_rank = players_filtered.sort_values(
    by="team_dependency_ratio",
    ascending=False
)

dependency_rank[[
    "player",
    "team",
    "goals",
    "goals_team_total",
    "team_dependency_ratio"
]].head(10)
```

```python
plt.figure()

top5_dep = dependency_rank.head(5)

plt.bar(top5_dep["player"], top5_dep["team_dependency_ratio"])
plt.xticks(rotation=45)
plt.xlabel("Player")
plt.ylabel("Team Goal Contribution Share")
plt.title("Top 5 Most Team-Dependent Players")
plt.show()
```

```python
players["impact_score"] = (
    players["goals_per_90"] +
    players["assists_per_90"]
)

impact_rank = players.sort_values(
    by="impact_score",
    ascending=False
)[["player", "team", "impact_score"]]

impact_rank.head(10)
```
