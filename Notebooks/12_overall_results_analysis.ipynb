```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
overall = pd.read_csv(
    r"../data/overwiev__results2024-202591_overall.csv"
)

overall.head()
```

```python
overall.columns
```

```python
overall = overall.rename(columns={
    "Squad": "team",
    "MP": "matches_played",
    "W": "wins",
    "D": "draws",
    "L": "losses",
    "GF": "goals_for",
    "GA": "goals_against",
    "GD": "goal_difference",
    "Pts": "points",
    "Pts/MP": "points_per_match",
    "Top Team Scorer": "top_scorer"
})

overall["team"] = overall["team"].str.strip()

overall.head()
```

```python
league_table = overall.sort_values(
    by="points",
    ascending=False
)

league_table[[
    "team",
    "wins",
    "draws",
    "losses",
    "points"
]]
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    league_table["team"],
    league_table["points"]
)

plt.xticks(rotation=45)
plt.ylabel("Points")
plt.title("Premier League Points Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    overall["xG"],
    overall["goals_for"]
)

plt.xlabel("Expected Goals (xG)")
plt.ylabel("Actual Goals Scored")
plt.title("Expected Goals vs Actual Goals")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    overall["xGA"],
    overall["goals_against"]
)

plt.xlabel("Expected Goals Against")
plt.ylabel("Actual Goals Conceded")
plt.title("Defensive Performance")

plt.show()
```

```python
overall[[
    "team",
    "top_scorer",
    "goals_for"
]].sort_values(
    by="goals_for",
    ascending=False
)
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    overall["team"],
    overall["Attendance"]
)

plt.xticks(rotation=45)
plt.ylabel("Average Attendance")
plt.title("Fan Attendance Across Teams")

plt.show()
```
