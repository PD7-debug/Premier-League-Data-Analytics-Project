```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
playing_time_for = pd.read_csv(
    r"../data/overwiev__stats_squads_playing_time_for.csv"
)

playing_time_for.head()
```

```python
playing_time_for.columns
```

```python
playing_time_for = playing_time_for.rename(columns={
    "Squad": "team",
    "# Pl": "players_used",
    "Playing_Time_MP": "matches_played",
    "Playing_Time_Min": "minutes_played",
    "Playing_Time_90s": "total_90s",
    "Subs_Subs": "subs_made",
    "Team Success_+/-": "goal_difference",
    "Team Success_+/-90": "goal_diff_per_90"
})

playing_time_for["team"] = playing_time_for["team"].str.strip()

playing_time_for.head()
```

```python
# Rotation Index
playing_time_for["rotation_index"] = (
    playing_time_for["players_used"] /
    playing_time_for["matches_played"]
)

# Substitution Intensity
playing_time_for["sub_intensity"] = (
    playing_time_for["subs_made"] /
    playing_time_for["matches_played"]
)

playing_time_for[[
    "team",
    "players_used",
    "rotation_index",
    "sub_intensity"
]].head()
```

```python
playing_time_for.columns.tolist()
```

```python
playing_time_for[[
    "team",
    "players_used",
    "rotation_index",
    "sub_intensity"
]].head()
```

```python
playing_time_for["minutes_concentration"] = (
    playing_time_for["Playing Time_Min"] /
    playing_time_for["players_used"]
)

playing_time_for[[
    "team",
    "players_used",
    "Playing Time_Min",
    "minutes_concentration"
]].head()
```

```python
concentration_rank = playing_time_for.sort_values(
    by="minutes_concentration",
    ascending=False
)

concentration_rank[[
    "team",
    "minutes_concentration"
]]
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    concentration_rank["team"],
    concentration_rank["minutes_concentration"]
)

plt.xticks(rotation=45)
plt.title("Minutes Concentration Across Teams")
plt.ylabel("Minutes per Player")

plt.show()
```

```python
playing_time_for[["team", "Age"]].head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    playing_time_for["Age"],
    playing_time_for["goal_diff_per_90"]
)

plt.xlabel("Average Squad Age")
plt.ylabel("Goal Difference per 90")
plt.title("Age vs Team Performance")

plt.show()
```

```python
sub_rank = playing_time_for.sort_values(
    by="sub_intensity",
    ascending=False
)

sub_rank[[
    "team",
    "sub_intensity"
]]
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    sub_rank["team"],
    sub_rank["sub_intensity"]
)

plt.xticks(rotation=45)
plt.ylabel("Substitutions per Match")
plt.title("Substitution Intensity Across Teams")

plt.show()
```
