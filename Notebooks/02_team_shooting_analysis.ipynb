```python
# Rename columns for easier use
shooting_for = shooting_for.rename(columns={
    "Squad": "team",
    "Standard_Gls": "goals",
    "Standard_Sh": "shots",
    "Standard_SoT": "shots_on_target",
    "90s": "matches_equivalent"
})

# Calculate metrics
shooting_for["shot_accuracy"] = shooting_for["shots_on_target"] / shooting_for["shots"]
shooting_for["conversion_rate"] = shooting_for["goals"] / shooting_for["shots"]
shooting_for["goals_per_90"] = shooting_for["goals"] / shooting_for["matches_equivalent"]

shooting_for[[
    "team",
    "goals",
    "shots",
    "shots_on_target",
    "shot_accuracy",
    "conversion_rate",
    "goals_per_90"
]].head()
```

```python
# Rename columns for easier use
shooting_for = shooting_for.rename(columns={
    "Squad": "team",
    "Standard_Gls": "goals",
    "Standard_Sh": "shots",
    "Standard_SoT": "shots_on_target",
    "90s": "matches_equivalent"
})

# Calculate metrics
shooting_for["shot_accuracy"] = shooting_for["shots_on_target"] / shooting_for["shots"]
shooting_for["conversion_rate"] = shooting_for["goals"] / shooting_for["shots"]
shooting_for["goals_per_90"] = shooting_for["goals"] / shooting_for["matches_equivalent"]

shooting_for[[
    "team",
    "goals",
    "shots",
    "shots_on_target",
    "shot_accuracy",
    "conversion_rate",
    "goals_per_90"
]].head()
```

```python
conversion_rank = shooting_for.sort_values(
    by="conversion_rate",
    ascending=False
)[["team", "goals", "shots", "conversion_rate"]]

conversion_rank.head(10)
```

```python
top_conversion = shooting_for.sort_values(
    by="conversion_rate",
    ascending=False
).head(10)

plt.figure()
plt.bar(top_conversion["team"], top_conversion["conversion_rate"])
plt.xticks(rotation=45)
plt.xlabel("Team")
plt.ylabel("Conversion Rate")
plt.title("Top 10 Teams by Shot Conversion Rate")
plt.show()
```

```python
plt.figure()
plt.scatter(shooting_for["shots"], shooting_for["conversion_rate"])
plt.xlabel("Total Shots")
plt.ylabel("Conversion Rate")
plt.title("Shot Volume vs Conversion Efficiency")
plt.show()
```

```python
shooting_for["efficiency_score"] = (
    shooting_for["shot_accuracy"] + shooting_for["conversion_rate"]
) / 2

efficiency_rank = shooting_for.sort_values(
    by="efficiency_score",
    ascending=False
)[["team", "efficiency_score"]]

efficiency_rank.head(10)
```

```python
team_combined = shooting_for.merge(
    standard_against[["team", "goals_conceded"]],
    on="team"
)

team_combined.head()
```

```python
# Normalize attack
team_combined["attack_index"] = (
    team_combined["goals"] / team_combined["goals"].max()
)

# Normalize defense (invert because lower conceded is better)
team_combined["defense_index"] = 1 - (
    team_combined["goals_conceded"] / team_combined["goals_conceded"].max()
)

# Composite score
team_combined["overall_strength_score"] = (
    team_combined["attack_index"] +
    team_combined["defense_index"]
) / 2

# Rank teams
strength_rank = team_combined.sort_values(
    by="overall_strength_score",
    ascending=False
)

strength_rank[["team", "overall_strength_score"]].head(10)
```
